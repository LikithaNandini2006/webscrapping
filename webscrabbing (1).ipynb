```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# Load dataset (CHANGE THIS PATH TO WHERE YOUR FILE IS LOCATED)
df = pd.read_csv(r"C:\\Users\\admin\\Desktop\\SSC GD\\webscrap.csv")

# Clean price column using regex to extract digits
df['price_num'] = df['Price'].apply(lambda x: int(re.sub(r'[^0-9]', '', str(x))))

# Create page numbers (40 products per page)
df['page'] = df.index // 40 + 1

# Select Page 1
page1 = df[df['page'] == 1]

# Scatter plot
plt.figure(figsize=(8,5))
plt.scatter(page1.index, page1['price_num'])
plt.xlabel("Product Index")
plt.ylabel("Price (₹)")
plt.title("Scatter Plot of Prices – Page 1")
plt.grid(True)
plt.show()
```

```python
Insight

Your dataset mainly uses:

“Only few left”

“Top Discount of the Sale”

→ No “out of stock” entries were found.
```

```python
import pandas as pd
import re

# Load dataset
df = pd.read_csv(r"C:\\Users\\admin\\Desktop\\SSC GD\\webscrap.csv")   # <-- change to your path

# Create clean discount percentage column
df['discount_pct'] = df['Discount'].apply(
    lambda x: int(re.sub(r'[^0-9]', '', str(x))) if pd.notnull(x) else 0
)

# Count products WITH discount
has_discount = df[df['discount_pct'] > 0].shape[0]

# Count products WITHOUT discount
no_discount = df[df['discount_pct'] == 0].shape[0]

print("Products with discount:", has_discount)
print("Products without discount:", no_discount)
```

```python
Insight

All products have a discount percentage, so:

100% products show discounts
```

```python
import pandas as pd
import re

# Load dataset
df = pd.read_csv(r"C:\\Users\\admin\\Desktop\\SSC GD\\webscrap.csv")   # <-- change to your correct path

# Clean price
df['price_num'] = df['Price'].apply(lambda x: int(re.sub(r'[^0-9]', '', str(x))))

# Clean discount percentage
df['discount_pct'] = df['Discount'].apply(lambda x: int(re.sub(r'[^0-9]', '', str(x))))

# Compute quartiles
q1 = df['price_num'].quantile(0.25)   # Entry-level
q3 = df['price_num'].quantile(0.75)   # Premium-level

# Entry-level = lowest 25%
entry_level = df[df['price_num'] <= q1]

# Premium = highest 25%
premium = df[df['price_num'] >= q3]

# Average discount comparison
entry_discount = entry_level['discount_pct'].mean()
premium_discount = premium['discount_pct'].mean()

print("Average discount for entry-level (cheapest 25%):", round(entry_discount, 2), "%")
print("Average discount for premium (top 25% price):", round(premium_discount, 2), "%")

if entry_discount > premium_discount:
    print("\n👉 Entry-level products ARE discounted more aggressively.")
else:
    print("\n👉 Premium products have equal or higher discounts.")
```

```python
import pandas as pd
import re

# Load dataset
df = pd.read_csv(r"C:\\Users\\admin\\Desktop\\SSC GD\\webscrap.csv")  # <-- replace with your correct path

# Verify rating column exists
if 'rating' not in df.columns:
    print("❌ Cannot compute brand loyalty premium — 'rating' column is missing.")
else:
    # Clean price
    df['price_num'] = df['Price'].apply(lambda x: int(re.sub(r'[^0-9]', '', str(x))))

    # Convert rating to numeric
    df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

    # Define top-rated brands = avg rating ≥ 4.0
    brand_rating = df.groupby('Brand')['rating'].mean()
    top_brands = brand_rating[brand_rating >= 4].index

    # Calculate price averages
    df['brand_type'] = df['Brand'].apply(lambda x: 'Top-rated' if x in top_brands else 'Regular')

    avg_prices = df.groupby('brand_type')['price_num'].mean()

    print("\nAverage Price by Brand Type:")
    print(avg_prices)

    if avg_prices['Top-rated'] > avg_prices['Regular']:
        print("\n👉 Top-rated brands command a price premium.")
    else:
        print("\n👉 Top-rated brands do NOT show a strong price premium.")
```

```python
Q9.For a given product, what percentage of reviews are 1–2 stars vs. 4–5 stars? give the code
```

```python
import pandas as pd

# Load dataset (change path)
df = pd.read_csv(r"C:\\Users\\admin\\Desktop\\SSC GD\\webscrap.csv")

# Required review columns
review_cols = ['1_star', '2_star', '3_star', '4_star', '5_star']

# Check if review columns exist
missing = [c for c in review_cols if c not in df.columns]

if missing:
    print("❌ Cannot compute review distribution — missing columns:", missing)
else:
    # Choose product (example: first product)
    product = df.iloc[0]

    # Total reviews
    total_reviews = product[review_cols].sum()

    # 1–2 star reviews
    low_reviews = product[['1_star', '2_star']].sum()
    low_pct = (low_reviews / total_reviews) * 100

    # 4–5 star reviews
    high_reviews = product[['4_star', '5_star']].sum()
    high_pct = (high_reviews / total_reviews) * 100

    print(f"Percentage of 1–2 star reviews: {low_pct:.2f}%")
    print(f"Percentage of 4–5 star reviews: {high_pct:.2f}%")
```

```python
Insight

When available, this metric tells you:

High 1–2 star % → quality issues

High 4–5 star % → customer satisfaction and strong product-market fit
```
